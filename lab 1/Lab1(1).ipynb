Write a method `gen_sinusoidal(N)` that generates toy data like in fig 1.2 of Bishop's book. The method should have a parameter $N$, and should return $N$-dimensional vectors $\bx$ and $\bt$, where $\bx$ contains evenly spaced values from 0 to (including) 2$\pi$, and the elements $t_i$ of $\bt$ are distributed according to:

$$t_i \sim \mathcal{N}(\mu_i, \sigma^2)$$

### 1.2 Polynomial regression (15 points)

Write a method `fit_polynomial(x, t, M)` that finds the maximum-likelihood solution of an _unregularized_ $M$-th order polynomial for some dataset `x`. The error function to minimize w.r.t. $\bw$ is:

$E(\bw) = \frac{1}{2} (\bPhi\bw - \bt)^T(\bPhi\bw - \bt)$

where $\bPhi$ is the _feature matrix_ (or _design matrix_) as explained in Bishop's book at section 3.1.1, $\bt$ is the vector of target values. Your method should return a vector $\bw$ with the maximum-likelihood parameter estimates. 

```python
'''
- takes a dataset(x, t) and finds the maximum likelihood solution of an unregularized M-th order polynomial
- checks if M > 0, and adds 1 to account for the case of M = 0
- creates the design matrix for the Mth order.
- uses the linalg.pinv() function, that directly creates the (Moore-Penrose) pseudo-inverse of the design matrix.
'''
def fit_polynomial(x, t, M):
    if M<0:
        return 'Please use an order greater than 0.'
    M += 1 #this is to allow for M = 0
    fi = np.zeros((len(x), M), dtype = int)

    for i in range(M):
        for j in range(len(x)):
            fi[j][i] = int(math.pow(x[j], i))
    
    w = np.dot(np.linalg.pinv(fi),t)
    
    return w

def Q_1_2():
    x, y = gen_sinusoidal(10)
    print(fit_polynomial(x, y, 2))
    
Q_1_2()
```

### 1.4 Regularized linear regression (10 points)

Write a method `fit_polynomial_reg(x, t, M, lamb)` that fits a _regularized_ $M$-th order polynomial to the sinusoidal data, as discussed in the lectures, where `lamb` is the regularization term _lambda_. (Note that 'lambda' cannot be used as a variable name in Python since it has a special meaning). The error function to minimize w.r.t. $\bw$:

$E(\bw) = \frac{1}{2} (\bPhi\bw - \bt)^T(\bPhi\bw - \bt) + \frac{\lambda}{2} \mathbf{w}^T \mathbf{w}$

For background, see section 3.1.4 of Bishop's book.

```python
'''
- takes a dataset(x, t) and finds the maximum likelihood solution of a regularized M-th order polynomial
- uses lamb = lambda as the regularization term
- checks if M >= 0, and adds 1 to account for the case of M = 0
- creates the design matrix for the Mth order.
- uses the linalg.pinv() function, that directly creates the (Moore-Penrose) pseudo-inverse of the design matrix.
'''
def fit_polynomial_reg(x, t, M, lamb):
    if M<0:
        return 'Please use an order greater than 0.'        
    M += 1 #this is to allow for M = 0
    fi = np.zeros((len(x), M), dtype = int)
    Ilamb = np.eye(M).dot(lamb)

    
    for i in range(M):
        for j in range(len(x)):
            fi[j][i] = int(math.pow(x[j], i))
    
    w = np.linalg.pinv(Ilamb + (fi.transpose().dot(fi))).dot(fi.transpose().dot(t))
    
    

    
    return w

def Q_1_4():
    x, y = gen_sinusoidal(10)
    print(fit_polynomial_reg(x, y, 3, -.5))
    
Q_1_4()
    
```

```python
def kfold_indices(N, k):
    all_indices = np.arange(N,dtype=int)
    np.random.shuffle(all_indices)
    idx = np.floor(np.linspace(0,N,k+1))
    train_folds = []
    valid_folds = []
    for fold in range(k):
        valid_indices = all_indices[idx[fold]:idx[fold+1]]
        valid_folds.append(valid_indices)
        train_folds.append(np.setdiff1d(all_indices, valid_indices))
        
    return train_folds, valid_folds

'''
- given a set of M, and a set of lambdas, find the best choice of M and lambda for a dataset(x,t)
- uses the fold, or leave-out crossvalidation technique
- finds the average error given a set of folds, and chooses the best average error for all M/lambda combinations
- plots findings in pretty_plot()
'''

def find_best_fit(M, lamb, dataset):
    
    #save the data points
    x, t = dataset
    
    #get training data indices
    training_x, valid_x = kfold_indices(9,9)

    #the choice of best M, l
    best_m = 0
    best_l = 0.00
    best_av_e = 100

    error_plot = []
    for m in M:
        for l in lamb:
    
            errors_to_average = []
            for i in range(len(training_x)):
    
                t_training_set = []
                x_training_set = []
                
                for k in training_x[i]:
                    t_training_set.append(t[k])
                    x_training_set.append(x[k])
        
                w = fit_polynomial_reg(x_training_set, t_training_set, m, l)
                
                t_valid_set = t[valid_x[i][0]]
                x_valid_set = [x[valid_x[i][0]]]
            
                calculated_fit = fitted_function(x_valid_set, w)
                
                sqrd_error = math.pow((t_valid_set - calculated_fit), 2)
                errors_to_average.append(sqrd_error)
   
            else:
                #average errors and save result
                e_mean = np.mean(errors_to_average)
                error_plot.append([m, l, e_mean])
                
                if(e_mean < best_av_e):
                    best_av_e = e_mean
                    best_m = m
                    best_l =l
    
    Ms = np.array(list(e[0] for e in error_plot))
    ls = np.array(list(e[1] for e in error_plot))
    errors = np.array(list(e[2] for e in error_plot))
    pretty_plot(x, t, Ms, ls, errors)
    
    return best_m, best_l

def pretty_plot(x, t, M, lamb, e):

    area = np.pi * (np.array(list(np.log(es) for es in e)))**2 
    colors = np.array(list(np.log(es) for es in e))

    fig = plt.figure(figsize=(10, 6), dpi=80)
    ax = plt.subplot(211)
    ax.scatter(M, lamb, s = [200 if c > np.min(e) - 0.001 and c < np.min(e) + 0.001 else 60 for c in e], c=colors, alpha = 0.5)
    plt.ylabel('$\lambda$')
    plt.xlabel('M')
    plt.title('Best choice highlighted by arrow')
    ax.annotate('M = %s, $\lambda$ = %s' % (M[np.argmin(e)], lamb[np.argmin(e)]), xy=(M[np.argmin(e)], lamb[np.argmin(e)]),  xycoords='data',
            xytext=(0.8, 0.95), textcoords='offset points',
            arrowprops=dict(facecolor='black', shrink=0.05, width = 1.0, headwidth = 8.0))
 
    #plot underfit, best, overfit
    min_e = np.argmin(e)
    
    under_w = fit_polynomial_reg(x, t, M[1], lamb[2])
    over_w = fit_polynomial_reg(x, t, M[len(M) - 1], lamb[8])
    best_w = fit_polynomial_reg(x, t, M[min_e], lamb[min_e]) 
    
    plt.subplot(234)
    plt.plot(x, fitted_function(x, under_w), color="green",  linewidth=1.0, linestyle="-")
    plt.plot(x,t,'bo')
    plt.title('M = ' + str(M[1]))
    plt.xlabel('x')
    plt.ylabel('t')
    plt.subplot(235)
    plt.plot(x, fitted_function(x, over_w), color="green",  linewidth=1.0, linestyle="-")
    plt.plot(x,t,'bo')
    plt.title('M = ' + str(M[len(M) - 1]))
    plt.xlabel('x')
    plt.ylabel('t')
    plt.subplot(236)
    plt.plot(x, fitted_function(x, best_w), color="green",  linewidth=1.0, linestyle="-")
    plt.plot(x,t,'bo')
    plt.title('M = ' + str(M[min_e]))
    plt.xlabel('x')
    plt.ylabel('t')
    plt.tight_layout()
    plt.show()

def Q_1_5():
    M = np.arange(0,10)
    lamb = []
    for i in range(11):
        lamb.append(np.exp(-10+i))

    M = np.linspace(min(M), max(M), 10, dtype = int)
    lamb.append(0)
    lamb = np.linspace(min(lamb), max(lamb), 10)    
    
    print('M' + str(M))
    print('l' + str(lamb))
    find_best_fit(M, lamb, gen_sinusoidal(9))

    
'''
- (fig.1) plots the errors for every pair of M and lambda, where blue is a low error and red a high one
- (fig.2) an example of underfitting, when the M is too low and thus constrains flexibility
- (fig.3) and example of overfitting, where the M is too high and the prediction starts to follow the noise
- (fig.4) the best fit found

'''    
Q_1_5()
    
```

### 2.2 Compute Posterior (15 points)

You're going to implement a Bayesian linear regression model, and fit it to the sinusoidal data. Your regression model has a zero-mean isotropic Gaussian prior over the parameters, governed by a single (scalar) precision parameter $\alpha$, i.e.:

$$p(\bw \;|\; \alpha) = \mathcal{N}(\bw \;|\; 0, \alpha^{-1} \bI)$$

The covariance and mean of the posterior are given by:

$$\bS_N= \left( \alpha \bI + \beta \bPhi^T \bPhi \right)^{-1} $$
$$\bm_N = \beta\; \bS_N \bPhi^T \bt$$

where $\alpha$ is the precision of the predictive distribution, and $\beta$ is the noise precision. 
See MLPR chapter 3.3 for background.

Write a method `fit_polynomial_bayes(x, t, M, alpha, beta)` that returns the mean $\bm_N$ and covariance $\bS_N$ of the posterior for a $M$-th order polynomial, given a dataset, where `x`, `t` and `M` have the same meaning as in question 1.2.

```python
"""
    - computes the covariance and mean of the posterior
    - adds 1 to account for the case of M = 0
    - creates a matrix Fi with X - M dimension to store the base function
    - computes and return the covariance and the mean with given formula
    - covariance is a M - M matrix, Mean is a 1 - M vector  
"""    

def fit_polynomial_bayes(x, t, M, alpha, beta):
    if M<=0:
        return 'Please use an order greater than 0.'
    M += 1 #this is to allow for M = 0
    fi = np.zeros((len(x), M), dtype = int)
    I = np.identity(M, dtype = float)
    alphaI = alpha * I

    for i in range(M):
        for j in range(len(x)):
            fi[j][i] = float(math.pow(x[j], i))
    
    cov = np.linalg.pinv(alphaI + beta*fi.transpose().dot(fi))
    mu = beta*cov.dot(fi.transpose()).dot(t)
    
    return cov, mu
    
def Q_2_2():
    print(fit_polynomial_bayes(gen_sinusoidal2(9)[0], gen_sinusoidal2(9)[1], 3, 0.1, 0.2))
```
